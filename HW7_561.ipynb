### Question 1

Let $b=f(a)$, let $\lambda=\nabla f(a)$, and let $\mu=\nabla g(f(a))$. If we define the following

1. $\phi(x)=f(x)-f(a)-\lambda(x-a)$


2. $\Phi(x)=g(y)-g(b)-\mu(y-b)$


3. $\rho(x) =g \circ f(x)-=g \circ f(a)-\mu \circ \lambda(x-a)$


Then

4. $\lim_{x \rightarrow a} \frac{|\phi(x)|}{|x-a|}=0$


5. $\lim_{y \rightarrow b} \frac{|\Phi(x)|}{|y-b|}=0$

and we must show that 

$\lim_{x \rightarrow a} \frac{|\rho(x)|}{|x-a|}=0$


Now

$
\begin{equation} 
\rho(x)=g(f(x))-g(b)-\mu(\lambda(x-a))
=g(f(x))-g(b)-\mu(f(x)-f(a)-\phi(x))
=[g(f(x))-g(b)-\mu(f(x)-f(a))]+\mu(\phi(x))
=\Phi(f(x))+\mu(\phi(x))
\end{equation}
$

Thus we must prove

6. $\lim_{x \rightarrow a} \frac{|\Phi(f((x))|}{|x-a|}=0$

7. $\lim_{x \rightarrow a} \frac{|\mu(\phi(x))|}{|x-a|}=0$


Equation 7. follows from eq4. If $\epsilon >0$ it follows from eq5. that for some $\delta>0$ we have

$|\Phi(f(x))|<\epsilon |f(x)-b| $ if $|f(x)-b|<\delta $

which is true if $|x-a|< \delta_1$, for a suitable $\delta_1$. Then

$
\begin{equation}
|\Phi(f(x))|<\epsilon|f(x)-b|=\epsilon|\phi(x)+\lambda(x-a)|\leq \epsilon|\phi(x)|+\epsilon M|x-a|
\end{equation}
$

for some M. Eq6 now follows.

Chain rule is therefore proved.
